#  Visualizing Earnings Based On College Majors

In this project, We explore the dataset on the job outcomes of students in USA who graduated from college between 2010 and 2012. 

Using the visualizations with Pandas, our goal is to answer the following questions:
+) Do students in more popular majors earn more money?
+) How many majors are predominantly male? Predominantly female?
+) Which category of majors have the most students?

Before creating data visualizations, we first import Pandas and Matplotlib libraries, open the data set and remove rows contanning null values. 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

recent_grads=pd.read_csv('recent-grads.csv')
recent_grads.iloc[0] # the first row
```

The we print the first and the last 5 rows of the dataframe to get to know our data.

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe() # summary for all numeric columns
```

```python
raw_data_count=len(recent_grads)
raw_data_count
```

Thus our data set has 173 rows. We now drop all the rows contaning null values since they will make trouble for matplotlib. 

By the program below, one can see that there is only one with missing values.

```python
recent_grads=recent_grads.dropna()
cleaned_data_count=len(recent_grads)
raw_data_count - cleaned_data_count
```

We start our visualizations by using scatter plots to explore the following relations of the data set:

```python
# Sample_size and Median earnings of full-time, year-round only
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', 
                  title='Sample size vs. Median', figsize=(6,6))
```

```python
# Sample size and Unemployment rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', 
                  title='Sample_size vs. Unemployment_rate', figsize=(6,6))
```

```python
# Full-time workers and Median earnings
recent_grads.plot(x='Full_time', y='Median', kind='scatter', 
                  title='Full-time vs. Median', figsize=(6,6))
```

```python
# Share women (women/total) and Unemployment rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', 
                  title='Share Women vs. Unemployment rate', figsize=(6,6))
```

```python
# Men graduated and Median earnings
recent_grads.plot(x='Men', y='Median', kind='scatter', 
                  title='Men vs. Median', figsize=(6,6))
```

```python
# Women graduated and Median earnings
recent_grads.plot(x='Women', y='Median', kind='scatter', 
                  title='Women vs. Median', figsize=(6,6))
```

From the above visualizations, we can counlude that:
+) More popular majors do not mean to earn more money (Sample size vs Median).
+) Majors with a small amount of full-time employees earn more money (Median and full-time).

In the next step, we use scatter matrix plot, the combination of both scatter plots and histograms, to visualize potential relations between columns and the distribution of individual columns on the same figure.

To do that, one need to import scatter_matrix from pandas.plotting module.

```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
